**Create a 2x2 grid of subplots using subplot(). In each subplot, plot different mathematical functions using plot(). Each subplot should represent a different mathematical function.**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

**In the first subplot (top-left), plot a sine wave (y = sin(x)) for x values between 0 and 2Ï€.**

```python
x = np.arange(0,2*np.pi,0.1)
y = np.sin(x)

fig, axs = plt.subplots(1,1, figsize=(5,6))

plt.plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='square')
plt.title('Sine Wave',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='upper right',facecolor='yellow',shadow=True,fontsize=9)
plt.show()
```

**In the second subplot (top-right), plot a cosine wave (y = cos(x)) for x values between 0 and 2Ï€.**

```python
x = np.arange(0,2*np.pi,0.1)
y = np.cos(x)

fig, axs = plt.subplots(1,1, figsize=(5,6))

plt.plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
plt.title('Cosine Wave',color='black')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='upper right',facecolor='yellow',shadow=True,fontsize=9)
plt.show()
```

**In the third subplot (bottom-left), plot a parabola (y = xÂ²) for x values between -5 and 5**

```python
x = np.arange(-5, 6, 1)  # Corrected: Use np.arange for integer steps
y = x**2

fig, axs = plt.subplots(1, 1, figsize=(5, 6))

plt.plot(x, y, color='green', marker='*', markersize=10, markeredgecolor='red', markerfacecolor='purple', linewidth=3.5, linestyle=':', label='parabola')
plt.title('Parabola', color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='upper right', facecolor='yellow', shadow=True, fontsize=9)
plt.show()
```

**In the fourth subplot (bottom-right), plot a linear function (y = 2x + 1) for x values between -5 and 5**

```python
x = np.arange(-5, 6, 1)
y = 2 * x + 1

fig, axs = plt.subplots(1, 1, figsize=(5, 6))

plt.plot(x, y, color='green', marker='*', markersize=10, markeredgecolor='red', markerfacecolor='purple', linewidth=3.5, linestyle=':', label='linear')
plt.title('Linear Function', color='black')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='upper right', facecolor='yellow', shadow=True, fontsize=9)
plt.show()
```
