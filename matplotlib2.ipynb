```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
df
```

**histogram**

```python
plt.hist(df['quality'])
plt.hist(df['quality'],color='red',edgecolor='black',orientation='vertical')

plt.title('alcohol')
plt.xlabel('wine quality')
plt.ylabel('frequency')

plt.show()
```

```python
plt.scatter(df['alcohol'],df['citric acid'],marker='o',color='blue')
plt.title('alcohol vs citric acid')
plt.xlabel('alcohol')
plt.ylabel('citric acid')
plt.axhline(0.25)
plt.axhline(0.75)
plt.axvline(8.9)
plt.axvline(10.5)
plt.show()


#plt.scatter(df_var['col1'],df_var['col2])
```
