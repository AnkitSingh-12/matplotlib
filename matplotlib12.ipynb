```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.arange(1,5)
y=x**2
plt.plot(x,y)
```

**color, title, ylabel, xlabel, legend**

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color='green')
plt.title('square b/w 1 to 5',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(['square'])
```

**marker,markersize,markeredge,markerfacecolor,linewidth,linestyle,legend,facecolor,show**


```python
x=np.array([1,2,3,4,5,6,7,9])
y=np.array([1,2,3,4,5,8,9,7])
z=np.array([1,6,4,3,7,8,4,6])
plt.plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
plt.plot(x,z,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='cube')
plt.title('square b/w 1 to 5',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc='upper right',facecolor='yellow',shadow=True,fontsize=9)
plt.show()
```

**subplot,figuresize,title,tightlayout**

```python
plt.figure(figsize=(10,10))
plt.subplot(3,3,1,facecolor='pink')
x=np.arange(1,5)
y=x**2
plt.plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
plt.title('plot1')
plt.xlabel('plot1')
plt.ylabel('plot1')

plt.subplot(3,3,5,facecolor='yellow')
x=np.arange(1,5)
y=x**2
plt.plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
plt.title('plot2')
plt.xlabel('plot2')
plt.ylabel('plot2')

plt.subplot(3,3,9,facecolor='purple')
x=np.arange(1,5)
y=x**2
plt.plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
plt.title('plot9')
plt.xlabel('plot9')
plt.ylabel('plot9')


plt.tight_layout()
plt.show()

#drawbacks= if we want to create grids  we have ton write subplot() 8 times to over come to this prob to should have to use subplots( )
```

**subplots**

```python
fig,axes=plt.subplots(nrows=2,ncols=3,figsize=(10,10))
x=np.arange(1,5)
y=x**2
axes[0][0].set_title('plot1')
axes[0][0].plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
axes[1][2].plot(x,y,color='green',marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
```

**figure,variable**

```python
fig=plt.figure(figsize=(10,10),facecolor='red')
fig.add_axes([0.1,0.1,0.8,0.8],facecolor='yellow')


var2=fig.add_axes([0.5,0.5,0.4,0.4],facecolor='green')
var2.plot(x,y,marker='*',markersize=10,markeredgecolor='red',markerfacecolor='purple',linewidth=3.5,linestyle=':',label='sqr')
var2.set_title('1st grid',loc='left')
var2.set_xlabel('x-axis')
var2.set_ylabel('y-axis')


var3=fig.add_axes([0.10,0.10,0.4,0.4],facecolor='pink')
var3.plot(x,y)
var3.set_title('2nd grid',loc='left')
var3.set_xlabel('x-axis')
var3.set_ylabel('y-axis')
```

**anoate**

```python
x=np.arange(1,6)
y=x**2
plt.plot(x,y,marker='o')
for i in range (len(x)):
    plt.annotate(f'({int(x[i])},{int(y[i])})',(x[i],y[i]),ha='left')

#f'{c[i],y[i]}-- it will goes to particular position
#(x[i],y[i])--it will print the partcular postion
#ha='center'--->horizontal aligment to print values in aligment--->ha='left' ha='right'
```

**axhline,axvline**

```python
x=np.arange(1,6)
y=x**2
plt.plot(x,y,marker='o')
for i in range (len(x)):
    plt.annotate(f'({int(x[i])},{int(y[i])})',(x[i],y[i]),ha='left')
plt.axhline(5)
plt.axhline(3,5)

plt.axvline(2,color='green')
plt.axvline(2,4)
```

**pie chart**


```python
plt.figure(figsize=(5,10))
plt.title('marks',color='red',fontsize='30')
val=[99,95,85,36,87]
lab=['python','java','data analysis','sql','web']
plt.pie(val,labels=lab,autopct='%0.2f%%',shadow=True,explode=(0.3,0,0.2,0,0),textprops={'fontsize':12})
plt.show()


#autopct=workflow-->value/sum of all the values*100

```

**doughnut piechart**

```python
plt.figure(figsize=(5,10))
plt.title('marks',color='red',fontsize='30',x=0.5,y=1.3)
col_var=['red','green','blue','yellow','yellow']
val=[99,95,85,36,87]
lab=['python','java','data analysis','sql','web']
plt.pie(val,labels=lab,autopct='%0.2f%%',colors=col_var,radius=1.5,pctdistance=0.8,textprops={'fontsize':12})
plt.pie([1],colors='white',radius=1)
plt.show()
```
